```python
def transformWavelet(data, wtype, directory):
    
    for i in range(len(data)):
        coeffs1 = pywt.dwt2(data[i], wtype)
        LL, (data) = coeffs1
        
    os.chdir(directory)
    
    img = plt.plot(data[0][:,:,2])
    
    j = 1
    while os.path.exists("%d.png" % j):
        j+=1
    
    cv2.imwrite("%d.png" % j, img)
```

```python
data = eval_img
wtype = "bior1.3"
directory = "/Users/ahmetuten/Desktop/archive/eval_graph"

transformWavelet(data, wtype, directory)
```

```python
def transformW(imgs, wtype, directory):
    
    os.chdir(directory)
    
    for i in tqdm(range(len(imgs))):
        coeffs1 = pywt.dwt2(imgs[i], 'bior1.3')
        LL, (data) = coeffs1
        standard = preprocessing.scale(data[0][:,:,2])
        plt.plot(standard)
        plt.savefig("%d.png" % i)
```

```python
imgs = eval_img
wtype = "bior1.3"
directory = "/Users/ahmetuten/Desktop/archive/eval_graph"

transformW(imgs, wtype, directory)
```
